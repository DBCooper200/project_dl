```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline

df = pd.read_csv("customer_churn.csv")
df.sample(5)
```

```python
df.drop('customerID',axis='columns',inplace=True)
```

```python
df.dtypes
```

```python
#Quick glance at above makes me realize that TotalCharges should be float but it is an object. Let's check what's going on with this column

df.TotalCharges.values
```

```python
#it is string. Lets convert it to numbers

pd.to_numeric(df.TotalCharges)
```

```python
#some values seems to be not numbers but blank string. Let's find out such rows

pd.to_numeric(df.TotalCharges,errors='coerce').isnull()
```

```python
df[pd.to_numeric(df.TotalCharges,errors='coerce').isnull()]
```

```python
df.shape
```

```python
df[df.TotalCharges!=' '].shape
```

```python
#Remove rows with space in TotalCharges

df1 = df[df.TotalCharges!=' ']
df1.shape
```

```python
df1.dtypes
```

```python
df1.TotalCharges = pd.to_numeric(df1.TotalCharges)
```

```python
df1[df1.Churn=='No']
```

```python
tenure_churn_no = df1[df1.Churn=='No'].tenure
tenure_churn_yes = df1[df1.Churn=='Yes'].tenure

plt.xlabel("tenure")
plt.ylabel("Number Of Customers")
plt.title("Customer Churn Prediction Visualiztion")


plt.hist([tenure_churn_yes, tenure_churn_no], rwidth=0.95, color=['pink','red'],label=['Churn=Yes','Churn=No'])
plt.legend()
```

```python
mc_churn_no = df1[df1.Churn=='No'].MonthlyCharges      
mc_churn_yes = df1[df1.Churn=='Yes'].MonthlyCharges      

plt.xlabel("Monthly Charges")
plt.ylabel("Number Of Customers")
plt.title("Customer Churn Prediction Visualiztion")


plt.hist([mc_churn_yes, mc_churn_no], rwidth=0.95, color=['pink','red'],label=['Churn=Yes','Churn=No'])
plt.legend()
```

```python
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}') 
```

```python
print_unique_col_values(df1)
```

```python
#Some of the columns have no internet service or no phone service, that can be replaced with a simple No

df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_columns:
    df1[col].replace({'Yes': 1,'No': 0},inplace=True)
```

```python
for col in df1:
    print(f'{col}: {df1[col].unique()}') 
```

```python
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
df1.gender.unique()
```

```python
df2 = pd.get_dummies(data=df1, columns=['InternetService','Contract','PaymentMethod'])
df2.columns
#one hot encoding
```

```python
df2.sample(5)
```

```python
df2.dtypes
```

```python
cols_to_scale = ['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
for col in df2:
    print(f'{col}: {df2[col].unique()}')
```
